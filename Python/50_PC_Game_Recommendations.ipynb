```python
import pandas as pd
import gazpacho
from gazpacho import Soup
from requests import get
```

```python
url = "https://www.boredpanda.com/best-pc-games/?utm_source=google&utm_medium=organic&utm_campaign=organic"
```

```python
resp = get(url)
resp.status_code
```

```python
games = Soup(resp.text)
```

```python
games
```

```python
games.find("div", {'class':'open-list-header'}, mode='first').strip() 
```

```python
titles = games.find("div", {'class': 'open-list-header'})

clean_titles = []

for title in titles:
    clean_titles.append(title.strip())
    
print(clean_titles)
```

```python
prefix = '#'
```

```python
new_titles = []
for item in clean_titles:
    words = item[item.find(prefix) + len(prefix):].split() # split the string into a list of words
    new_item = ' '.join(words[1:]) # join the words back together, skipping the first word (numbering)
    new_titles.append(new_item)

print(new_titles)
```

```python
games_pc = pd.DataFrame({
    "title": new_titles
})

games_pc.head(10)
```
